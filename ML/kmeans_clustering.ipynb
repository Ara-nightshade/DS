```python
# Fearture scaling
dataa=data1.iloc[:,2:6].values 
#strandardize thee data to normal distribution
from sklearn import preprocessing
data1_stand=preprocessing.scale(dataa)
data1_stand=pd.DataFrame(data1_stand)
print(data1_stand)
```

```python
# wss plot 
# is used to find appropriate cluster number
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,8))
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(data1_stand)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('The elbow method')
plt.show()

#hit and try method to find the elbow... #here its 3
```

```python
# k means clustering and cluster information
#fitting k means to the dataset
# Assuming data1_stand is your standardized data
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(data1_stand)

#begining of the cluster numbering with 1 instead 0
y_kmeans1=y_kmeans
y_kmeans1=y_kmeans+1
cluster=pd.DataFrame(y_kmeans1)
data1['cluster']=cluster
```

## K nearest neighbor

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris=load_iris()
iris.data#features
iris.target#output
```

```python
irisdf=pd.DataFrame(data=iris.data, columns=iris.feature_names)
irisdf['class']=iris.target
irisdf.head()
```

```python
#splitting 
#x=irisdf.iloc[:,:-1].values
#y=irisdf.iloc[:,4].values

#here v its already split
x=iris.data
y=iris.target
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=.8)
```

```python
#pre-processing and standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(xtrain)
xtrain=scaler.transform(xtrain)
xtest=scaler.fit_transform(xtest)
```

```python
# K-nearest neighbours classifier
from sklearn.neighbors import KNeighborsClassifier

classifier=KNeighborsClassifier(n_neighbors=13)
classifier.fit(xtrain,ytrain)
```

```python
#Predictions and Evaluations
ypred=classifier.predict(xtest)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
```

```python
# Detection of optimal vwalue for K Neighnours 
```

```python
error=[]
#calculating error for K values between 1 and 40
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    pred_i=knn.predict(xtest)
    error.append(np.mean(pred_i != ytest))
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,40),error, color='red',linestyle='dashed',marker='d',markerfacecolor='blue',markersize=10)
plt.title('Error Rate K value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
plt.show()
```
