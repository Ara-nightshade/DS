```python
##Anova test
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
mod= ols('weight ~ group',#ols is used in regression model
         data=data).fit()
aov_table=sm.stats.anova_lm(mod,typ=2)#here typ=2 for more than 1 sample ... typ=1 for 1 sample .. normal z and t test is done #here we are checking what are the changes in weight with resprct to group
print(aov_table)#'weight'--> dependent variable.. target variable , 'group'-- independent variable predictor variable
```

```python
#from the table - F value isnt close to one- hence the variences are different hence 
#the p value say tat v need to regect the hypothesis , hence means aren't same for any teatment

#we have 3 checks -- f-test -- box plot(large data may lead to clutter) -- p value
#we can pass alpha value seperatly... default value is 95%
```

## Tukey's test

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison

mc=MultiComparison(data.weight,data.group)#it does comparision between different groups
#it gives true for the combination having highest difference in there variences
result=mc.tukeyhsd()

print(result)
print(mc.groupsunique)

```
