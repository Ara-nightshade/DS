```python
import numpy as np
import random

#function for rolling dice
def roll_the_dice(times):
    count=0
    for i in range(times):
        dice1=random.randint(1,6)
        dice2=random.randint(1,6)
        
        summ=dice1+dice2
        
        if summ>7 or ((summ%2)==0):
            count+=1
    return count/times


times=int(input("enter the number of times the dice must be rolled "))
result=np.round(roll_the_dice(times),2)
print(f"the probability of rolling the dices with even sum or greater than 7 is {result} ")       
```

```python
#A box contains 12 balls of type A and 18 of type B
#one transistor is taken out at random and returned 
#This process is repeated
#determine the probability that the first chosen is type A and second is type B

import random 
trials=10000
totalA=12
totalB=18

success=0

choicelist=list('A'*totalA + "B"*totalB)

def balls():
    return random.choice(choicelist)

for i in range(trials):
    try1=balls()
    try2=balls()
    if try1 =="A" and try2 =="B":
        success+=1
print(float(success)/trials)
```

- **Maximum Height:** The highest point of the curve is called the maximum. It's the point where the curve is the tallest, and it represents the most likely value in your data.

- **Half of Maximum Height:** Now, imagine drawing a horizontal line at exactly half the height of the maximum point. This line will intersect the curve at two points on each side of the peak.

- **Full Width at Half Maximum (FWHM):** The FWHM is the distance between these two points where the curve intersects the half-height line. It's the width of the curve when it's "half" as tall as the maximum point.

In simple terms, the FWHM tells you how "wide" the bell-shaped curve is when you measure it at the height where it's exactly halfway between the peak and the baseline. This width is an important parameter because it gives you an idea of how spread out the data is.
```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
plt.ylabel('Probability of car passing')
plt.xlabel('number of cars')
plt.title("probability distribution curve")

arr=[]
rv=poisson(25)#rate lambda value
for num in range(0,40):
    arr.append(rv.pmf(num))#pmf-->probability mass function p(x).... for x value from 1 to 40 p(x) is generated
    
prob=rv.pmf(28)#probabilty for x=28
print(prob)
plt.grid(True)
plt.plot(arr, linewidth=2.0)
plt.plot([28],[prob],marker='o',markersize=6,color='red')
plt.show()
```
